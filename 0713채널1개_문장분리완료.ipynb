```python
import openpyxl
wb2 = openpyxl.Workbook()
s = wb2.active
s.append(["인덱스",'원글 번호','내용'])
```

```python
num = 0
length = len(df2v)
length
```

```python
for i in range(0,length):
    dfsent = df2v['내용'][i].replace('\n','#').replace('1.','#').replace('2.','#').replace('3.','#').replace('4.','#').split('#')
    # 줄바꿈, 1.2.3.4 ... 등등 번호 붙인거 분리
    for j in range(0,len(dfsent)):
        go = dfsent[j]
        if (dfsent[j]==""): #공백이면
            continue
        num+=1
        s.append([num,i,go])
    
#    print(DataFrame(dfsent))

#split = split.apply(lambda x: pd.Series(x))
#split.stack().reset_index(level=1, drop=True).to_frame('after split')
```
